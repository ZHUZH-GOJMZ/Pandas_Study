## 一、单级索引

### 1、loc方法、iloc方法、[ ]操作符

最常用的索引方法可能就是这三类，其中iloc表示位置索引，loc表示标签索引，[ ]也具有很大的便利性。

下面我们来一一介绍。

#### 1） loc 方法

注意：所有在loc中使用的切片全部包含右端点！这是因为如果作为Pandas的使用者，那么肯定不太关心最后一个标签再往后一位是什么，但是如果是左闭右开，那么就很麻烦，先要知道再后面一列的名字是什么，非常不方便，因此Pandas中将loc设计为左右全闭

```python
df.loc[1105]  # 单行索引
```

```python
df.loc[[1102, 1104, 1105]]  # 多行索引
# 需要用[ ]括起来这几行ID，才能返回要索引的这几行。
```

```python
df.loc[1103:].head()  # 用切片形式的多行索引，这里是Python3切片的应用了
```

```python
df.loc[2401::-1].head()  # 也是切片的应用，这里[::-1]是Python倒序切片方法，所以我们看到ID的排列是倒序的。
```

```python
df.loc[:,'Height'].head()  # 单列索引
```

```python
df.loc[:,['Height', 'Weight']].head()  # 多列索引，跟多行索引是一样的，需要用[ ]括起需要返回的列名。
```

```python
df.loc[:,'Address':'Math'].head()  # 多列索引应用。这里我们如果用切片来返回多列，就可以不用[ ]，直接写头尾的列名即可。
# 这里还要注意一点，输出表包含了右端点 'Math' 的这一列的，也就是 loc 用切片包含右端点，这个和 iloc 是不一样的，后面我们会讲到。
```

```python
df.loc[:,'Math':'Address':-1].head()  # 多列索引应用，这里是倒序返回的举例。
```

```python
df.loc[1102:2401:3,'Address':'Math'].head()  # 联合索引
```

* 这里我们总结 loc[ ] 的格式用法： 
**loc [ ( 行变化 ) , ( 列变化 ) ]**    ，这里的行变化和列变化可以写单个行列名，也可以通过切片形式来写。

```python
df.loc[lambda x:x['Gender']=='M'].head()  
# 函数式索引，loc中使用的函数，传入参数就是前面的df，最后返回指定列中满足条件的行。
```

```python
def f(x):
    return [1101,1103]
df.loc[f]
# 这里的例子表示，loc中能够传入函数，并且函数的输入值是整张表，输出为标量、切片、合法列表（元素出现在索引中）、合法索引
```

```python
df.loc[df['Address'].isin(['street_7','street_5'])].head()  # 一般索引，索引指定列中的条件满足多个条件的行。
```

```python
df.loc[[True if i[-1]=='4' or i[-1]=='7' else False for i in df['Address'].values]].head()
# 布尔索引。这里通过 for 循环来遍历 Address 列，然后 True 代表返回，False 代表不返回。
# 这里只要 i[-1] 也就是地址名的最后一个字符数字，我们要求是 4 和 7 ，这样得到的结果会是和上一个代码一样。
# 也就是我们可以通过布尔索引特别轻松的来完成了很多所需要求。
```

本质上说，loc中能传入的只有布尔列表和索引子集构成的列表，只要把握这个原则就很容易理解上面那些操作

#### 2）iloc 方法

注意与loc不同，切片右端点不包含

```python
df.iloc[4]  # 单行索引，我们发现使用 iloc 是需要我们直接在 [ ] 中写所需要返回的行位置，而不是行名。
```

```python
df.iloc[3:5]  # 多行索引，用切片进行。
# 这里能够发现位置为 5 的数据并没有返回，这里就体现了刚才提到的，iloc 切片的右端点是不被包含的，也就是不会被返回。
```

```python
df.iloc[:,3].head()  # 单列索引，和 loc 类似，就是这里是输入列位置。
```

```python
df.iloc[:,7::-2].head()  # 多列索引，这个切片的意思是：从位置 7 开始，倒序隔一列返回一次。
```

```python
df.iloc[3::4,7::-2].head()  # 混合索引。
```

```python
df.iloc[lambda x:[3,7]].head()  # 函数式索引，只返回位置 3 和 7 的这两行。
```

iloc 中接收的参数只能为整数或整数列表，也可以使用布尔列表，但**不能使用纯布尔Series索引**，如果要用就必须像下面这样把values拿出来。

```python
#df.iloc[df['School']=='S_1'].head() #报错
df.iloc[(df['School']=='S_1').values].head()
```

* 这里小结一下 loc 和 iloc 的区别：

loc [ ] 中写的是行列名、函数或布尔索引， iloc [ ] 中写的是行列位置，函数，不能用布尔索引。

###  2、布尔索引

#### 布尔符号：'&','|','~'：分别代表和and，或or，取反not

```python
df[(df['Gender']=='F')&(df['Address']=='street_2')].head()  
# 返回同时满足这两个条件的行
```

```python
df[(df['Math']>85)|(df['Address']=='street_7')].head()  
# 返回任意满足其一条件的行
```

```python
df[~((df['Math']>75)|(df['Address']=='street_2'))].head()  
# 返回不满足上述任意一个条件的行。
```

这里就考验我们的对于“与或非”的应用了，很容易出错，需要注意。

#### loc和[ ]中相应位置都能使用布尔列表选择:

```python
df.loc[df['Math']>60,df.columns=='Physics'].head()
```

```python
df.loc[df['Math']>60,(df[:8]['Address']=='street_6').values].head()
# df.loc[df['Math']>60,(df[:8]['Address']=='street_6')].head() 如果去掉values会报错
# 这个得到的结果和上面是一样的。
```

#### isin 方法

```python
df[df['Address'].isin(['street_1','street_4'])&df['Physics'].isin(['A','A+'])]
# 可见 isin 方法可以更深入的查找，针对对应的列中继续查找满足条件的元素对应行。
# 这里也是 isin 方法和前面的 & 方法结合查找。
```

```python
# 上面的表也可以用字典方式来写：
df[df[['Address','Physics']].isin({'Address':['street_1','street_4'],'Physics':['A','A+']}).all(1)]
# all 与 & 的思路是类似的，其中的1代表按照跨列方向判断是否全为True
# 这里要注意括号的区别，容易出错。
```

### 3、快速标量索引

当我们只需要取一个元素时，at 和 iat 方法能够提供更快的实现：

```python
display(df.at[1101,'School'])
display(df.loc[1101,'School'])
display(df.iat[0,0])
display(df.iloc[0,0])
# 上面这些写法返回结果是一样的。
```

### 4、区间索引
此处介绍并不是说只能在单级索引中使用区间索引，只是作为一种特殊类型的索引方式，在此处先行介绍。

#### 利用interval_range方法

```python
pd.interval_range(start=0,end=5)
# closed参数可选'left''right''both''neither'，默认左开右闭。
```

```python
pd.interval_range(start=0,periods=8,freq=5)
# start参数控制起始元素，periods参数控制区间个数，freq控制步长
```

#### 利用cut将数值列转为区间为元素的分类变量，例如统计数学成绩的区间情况：

```python
math_interval = pd.cut(df['Math'], bins=[0,40,60,80,100])
# 注意，如果没有类型转换，此时并不是区间类型，而是category类型
# 这里将元素分成四个区间，下面的元素返回的是对应的区间。
math_interval.head()
```

#### 区间索引的选取

```python
df_i = df.join(math_interval,rsuffix='_interval')[['Math','Math_interval']].reset_index().set_index('Math_interval')
# 用 join 建立一列‘Math_interval’，因为 math_interval.name 为‘Math’，右后缀 rsuffix = ‘_interval’。
# 提取出 Math 和 Math_interval 这两列，然后重新设置索引名为 Math_interval 。
df_i.head()
```

```python
df_i.loc[65].head()
# 只要包含这个 65 的所有区间都会被返回。
```

```python
df_i.loc[[70,85]]
# 返回包含 70 或 85 的区间，但这里是先返回了 70 的区间，再返回了 85 的区间。
```

#### 如果想要选取某个区间，先要把分类变量转为区间变量，再使用overlap方法：

```python
# df_i.loc[pd.Interval(70,75)].head() 会报错
df_i[df_i.index.astype('interval').overlaps(pd.Interval(70,85))].head()
# 这里是按原顺序对应返回相应区间。
```

#### 索引Slice可以与loc一起完成切片操作，主要有两种用法:
#### 1）loc[idx[\*,\*]]型
第一个星号表示行，第二个表示列，且使用布尔索引时，需要索引对齐

```python
df_s.loc[idx['B':,df_s.iloc[0]>0.6]]
# 第一行元素中 大于0.6的返回。
# df_s.loc[idx['B':,df_s.iloc[:,0]>0.6]]  索引没有对齐报错
```

```python
df_s.loc[idx[df_s.iloc[:,0]>0.6,:('E','f')]]
# 这里就是第一列中大于0.6的元素返回，然后也要在E f之前。
```

#### 2）loc[idx[\*,\*],idx[\*,\*]]型
这里与上面的区别在于（a）中的loc是没有逗号隔开的，但（b）是用逗号隔开，前面一个idx表示行索引，后面一个idx为列索引

这种用法非常灵活，因此多举几个例子方便理解

```python
df_s.loc[idx['A'],idx['D':]]
# 后面的层出现，则前面的层必须出现
# df_s.loc[idx['a'],idx['D':]] #报错
```

```python
df_s.loc[idx[:'B','b':],:] 
# 举这个例子是为了说明①可以在相应level使用切片②某一个idx可以用:代替表示全选
```

```python
df_s.iloc[:,0]>0.6  # 返回布尔索引
```

```python
df_s.loc[idx[:'B',df_s.iloc[:,0]>0.6],:] 
# 这个例子表示相应位置还可以使用布尔索引
```

```python
# 特别要注意，（b）中的布尔索引是可以索引不对齐的，只需要长度一样，比如下面这个例子
df_s.loc[idx[:'B',(df_s.iloc[0]>0.6)[:6]],:]
```

```python
df_s.loc[idx[:'B','c':,(df_s.iloc[:,0]>0.6)],:]
# idx中层数k1大于df层数k2时，idx前k2个参数若相应位置是元素或者元素切片，则表示相应df层的元素筛选，同时也可以选择用同长度bool序列
# idx后面多出来的参数只能选择同bool序列，这样设计的目的是可以将元素筛选和条件筛选同时运用
```

```python
df_s.loc[idx[:'B',(df_s.iloc[:,0]>0.6),(df_s.iloc[:,0]>0.6)],:] 
# 这个就不是元素筛选而是条件筛选
# df_s.loc[idx[:'B',(df_s.iloc[:,0]>0.6),'c',:]] # 小索引的位置问题 报错
# df_s.loc[idx[:'c','B',(df_s.iloc[:,0]>0.6),:]] 小索引在大索引前 报错
```

#### （d）请问六项种族指标（HP、物攻、特攻、物防、特防、速度）极差的均值最大的是哪个属性（只考虑第一属性，且均值是对属性而言）？

```python
df['range'] = df.loc[:,'HP':'Speed'].max(axis=1) - df.loc[:,'HP':'Speed'].min(axis=1)
# 计算极差放入DataFrame中
datafra = df[['Type 1','range']].set_index('Type 1')
# 新建datafra保存Type 1和range列，并且将Type 1设置为索引。
max_range = 0
result = ''
# 然后我们通过for循环来计算均值。
for i in datafra.index.unique()[:-1]:
    temp = datafra.loc[i,:].mean()  # 求每个属性的种族指标均值
    if temp[0] > max_range:
        max_range = temp.values[0]  # 存放最大均值
        result = i  # 存放最大均值对应的属性
result
```

所以 Psychic 属性占比最高。

```python
df_to = df.query('Legendary==True')[['Type 1','Total']].set_index('Type 1')
# 建立新df_to保存神兽中的Type 1和Total，并且将Type 1 设置为索引。
max_range = 0
result = ''
# 循环求均值 和上一题一模一样。
for i in df_to.index.unique()[:-1]:
    temp = df_to.loc[i,:].mean()
    if temp[0] > max_range:
        max_range = temp.values[0]
        result = i
result
```

这里 Normal 属性神兽的种族值是最高的，而不是 Psychic 。
