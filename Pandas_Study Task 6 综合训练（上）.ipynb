## 一、2002 年-2018 年上海机动车拍照拍卖

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('joyful-pandas-master/2002年-2018年上海机动车拍照拍卖.csv')
df.head()
```

### (1) 哪一次拍卖的中标率首次小于 5%？

```python
df['gain rate']=df['Total number of license issued']/df['Total number of applicants']
df
```

```python
df_a=df[df['gain rate']<0.05]
df_a.head()
```

```python
df_a['Date'].iloc[0]
```

### (2) 按年统计拍卖最低价的下列统计量：最大值、均值、0.75 分位数，要求显示在同一张表上。

```python
df_d = df.groupby('Year')['lowest price'].agg(['max','mean',lambda x:x.quantile(0.75)])
df_d.rename(columns={'<lambda_0>':'0.75quantile'})
```

### (3) 将第一列时间列拆分成两个列，一列为年份（格式为 20××），另一列为月份（英语缩写），添加到列表作为第一第二列，并将原表第一列删除，其他列依次向后顺延。

```python
df['Year']=df['Date'].map(lambda x: int(x.split('-')[0]))+2000
df['Month']=df['Date'].map(lambda x: x.split('-')[1])
df
```

```python
df_b=df.drop(columns='Date') # 取出Date列
df_b.head()
```

```python
cols=list(df_b)  # 获取列标签
cols.insert(0,cols.pop(cols.index('Year')))
cols.insert(1,cols.pop(cols.index('Month')))
cols
```

```python
df_c = df_b.loc[:,cols]
df_c.head()
```

### (4) 现在将表格行索引设为多级索引，外层为年份，内层为原表格第二至第五列的变量名，列索引为月份。

```python
del df_c['gain rate']
df_c.set_index(['Year']+cols[2:6])
```

### (5) 一般而言某个月最低价与上月最低价的差额，会与该月均值与上月均值的差额具有相同的正负号，哪些拍卖时间不具有这个特点？

```python
df_e = df_c[['Year','Month','lowest price','avg price']].set_index(['Year','Month'])
df_e.head()
```

```python
df_e[df_e.diff().iloc[:,0]/df_e.diff().iloc[:,1] < 0]
```

### (6) 将某一个月牌照发行量与其前两个月发行量均值的差额定义为发行增益，最初的两个月用 0 填充，求发行增益极值出现的时间。

```python
df_g = df_c['Total number of license issued']-df_c['Total number of license issued'].rolling(2).mean().shift(1)
df_c['gain']=df_g.fillna(0)
df_c.head()
```

```python
print("极大值为：", df_c['gain'].idxmax())
print("极小值为：", df_c['gain'].idxmin())
```

### (4) 若某机场至少存在 5 年或以上满足所有月运量记录都为 0，则将其所有年份的记录信息从表中删除，并返回处理后的表格

```python
df_b = df[(df.loc[:,'January':'December']==0).all(axis=1)]
df_b.head()
```

```python
airp = df_b.groupby('Airport name')['Year'].count()
airp
```

```python
realairp = airp[airp>5].index.tolist()
df_c = df[~df['Airport name'].isin(realairp)]
df_c.head()
```

### (5) 采用一种合理的方式将所有机场划分为东南西北四个分区，并给出 2017年-2019 年货运总量最大的区域。

### (6) 在统计学中常常用秩代表排名，现在规定某个机场某年某个月的秩为该机场该月在当年所有月份中货运量的排名（例如 *** 机场 19 年 1 月运量在整个 19 年 12 个月中排名第一，则秩为 1），那么判断某月运量情况的相对大小的秩方法为将所有机场在该月的秩排名相加，并将这个量定义为每一个月的秩综合指数，请根据上述定义计算 2016 年 12 个月的秩综合指数。

```python
df_b = df.iloc[:,:-2].query('Year == 2016').copy()
df_b.update(df_b.loc[:,'January':'December'].apply(lambda x:x.rank(method = 'min'),axis = 1))
df_b.head()
```

```python
df_b.set_index('Airport name',inplace=True)
del df_b['Year']
df_b.sum()
```
